```python
import nnfs
from nnfs.datasets import spiral_data

nnfs.init()   # sets default seed and dtype
X, y = spiral_data(samples=100, classes=3)
```

```python
class Layer_Dense:
  def __init__(self, features, neurons):
    self.weights = 0.01 * np.random.randn(neurons, features)
    self.bias = np.zeros(neurons)

  def forward(self, inputs):
    self.output = np.dot(inputs, self.weights.T) + self.bias
```

```python
## Create the first dense layer
dense1 = Layer_Dense(2, 3)
dense1.forward(X)
print(f"The shape of the output is : {dense1.output.shape}")
```

```python
## Create the second dense layer
dense2 = Layer_Dense(3, 10)
dense2.forward(dense1.output)
print(f"The shape of the next output layer is : {dense2.output.shape}")
```

```python
class Activation_RelU:
  def forward(self, inputs):
    self.output = np.maximum(0, inputs)
```

```python
inputs = np.array([[1, 2, 3, 2.5],
                   [2., 5., -1., 2],
                   [-1.5, 2.7, 3.3, -0.8]])
```

```python
exp_values = np.exp(inputs - np.max(inputs, axis = 1, keepdims = True))
probabilities = exp_values / np.sum(exp_values, axis = 1, keepdims = True)
print(probabilities)
```

```python
np.sum(probabilities, axis = 1)
```

```python
class Activation_Softmax:
  def forward(self, inputs):
    exp_values = np.exp(inputs - np.max(inputs, axis = 1, keepdims = True))
    probabilities = exp_values / np.sum(exp_values, axis = 1, keepdims = True)
    self.output = probabilities
```

```python
## Create a Dataset
X, y = spiral_data(samples=100, classes=3)
print(f"Shape of the input : {X.shape}")
## Create the first dense layer
dense1 = Layer_Dense(2, 3)
## Forward pass
dense1.forward(X)
## Get output of the first result , output size will be 300 * 3
output1 = dense1.output
print(f"Shape of the first output : {output1.shape}")
## Apply RelU Activation function on the output
activation1 = Activation_RelU()
activation1.forward(output1)
output1 = activation1.output
## Create the second dense layer
dense2 = Layer_Dense(3, 3)
## Forward pass
dense2.forward(output1)
## Get output of the second result, output size will be 300 * 3
output2 = dense2.output
print(f"Shape of the second output : {output2.shape}")
## Apply Softmax Activation on the output
activation2 = Activation_Softmax()
activation2.forward(output2)
output2 = activation2.output
## Print the output
print(f"The shape of the final output : {output2.shape}")
print(output2)
```
